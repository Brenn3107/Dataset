## Data loading

### Subtask:
Load the "TeleCom.xlsx" file into a pandas DataFrame.


**Reasoning**:
Load the excel file into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    df = pd.read_excel('TeleCom.xlsx')
    display(df.head())
except FileNotFoundError:
    print("Error: 'TeleCom.xlsx' not found. Please ensure the file is in the correct location.")
    df = None
except Exception as e:
    print(f"An error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its basic characteristics.


**Reasoning**:
I need to explore the dataset's characteristics, including shape, data types, descriptive statistics, and value counts for categorical columns.  This can all be done in a single code block.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Check data types
print("\nData Types:\n", df.dtypes)

# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:\n", df.describe())

# Value counts for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nValue Counts for {col}:\n{df[col].value_counts()}")
```

## Data cleaning

### Subtask:
Clean the dataset by handling missing values and outliers.


**Reasoning**:
Identify and handle missing values and outliers in the dataset.



```python
# Check for missing values
print(df.isnull().sum())

# Impute missing values for numerical features (if any)
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Impute missing values for categorical features (if any)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

# Outlier handling
import matplotlib.pyplot as plt
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    df.boxplot(column=[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

    # Example outlier handling: Winsorizing
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Check for inconsistencies and standardize values in categorical columns
for col in categorical_cols:
    print(f'Unique values in {col}: {df[col].unique()}')
    # Example: Standardize 'education' column
    if col == 'education':
        df['education'] = df['education'].replace({'basic.4y':'basic', 'basic.6y':'basic', 'basic.9y':'basic'})
```

## Data visualization

### Subtask:
Visualize the cleaned data to gain insights into the distribution of variables and relationships between them.


**Reasoning**:
Visualize the cleaned data using histograms, box plots, count plots, scatter plots, and a correlation heatmap to gain insights into the distribution of variables and relationships between them.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Box plots for numerical features
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Count plots for categorical features
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[col])
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Scatter plots for pairs of numerical features
sns.pairplot(df, vars=numerical_cols[:5], hue='y') #limit to first 5 numerical features for better visualization, adjust as needed
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## Data analysis

### Subtask:
Perform descriptive statistics and correlation analysis on the cleaned dataset.


**Reasoning**:
Calculate descriptive statistics, correlation matrix, and analyze the relationship between numerical and categorical features, as well as the target variable distribution.



```python
# Descriptive statistics for numerical features
numerical_features = df.select_dtypes(include=['number']).columns
descriptive_stats = df[numerical_features].describe()
display(descriptive_stats)

# Correlation matrix
correlation_matrix = df[numerical_features].corr()
display(correlation_matrix)

# Identify highly correlated features (threshold: 0.7 or -0.7)
threshold = 0.7
highly_correlated_features = correlation_matrix[abs(correlation_matrix) > threshold]
display(highly_correlated_features)

# Analyze relationship between numerical and categorical features (example: average age per job)
average_age_per_job = df.groupby('job')['age'].mean()
display(average_age_per_job)

# Analyze target variable distribution
target_distribution = df['y'].value_counts(normalize=True) * 100
display(target_distribution)
```
